```python

import pandas as pd
import numpy as np


file_path = '../dynamic_pricing.csv'  

try:
    df = pd.read_csv(file_path)
    print("File loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}. Please check the file path.")

print("First few rows of the data:")
print(df.head())

print("\nMissing values in each column:")
print(df.isnull().sum())

#   missing values separately for numerical and categorical columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())


categorical_columns = df.select_dtypes(include=[object]).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

# Exploratory Data Analysis (EDA)
# Displaying summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(df.describe())

# Displaying basic info about the dataset (column types, non-null counts)
print("\nInformation about the dataset:")
print(df.info())

# Checking  if there are any duplicate rows
print("\nCheck for duplicate rows in the dataset:")
print(f"Duplicate rows: {df.duplicated().sum()}")

 
#  save the cleaned dataset as a new CSV file
df.to_csv('../cleaned_dynamic_pricing.csv', index=False)

#  Display the cleaned data head again to verify
print("\nCleaned data (first few rows):")
print(df.head())
```
